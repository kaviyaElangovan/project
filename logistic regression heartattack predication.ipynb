```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/rashida048/Datasets/refs/heads/master/Heart.csv')
```

```python
df
```

```python
df.head()
```

```python
df['ChestPain']=df['ChestPain'].astype("category")
df['ChestPain']=df['ChestPain'].cat.codes
df
```

```python
df['Thal']=df['Thal'].astype("category")
df['Thal']=df['Thal'].cat.codes
df
```

```python
df['AHD']=df['AHD'].astype("category")
df['AHD']=df['AHD'].cat.codes
df
```

```python
df=df.drop(columns="Unnamed: 0")
df
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df=df.dropna()
df
```

```python
df.corr()
```

```python
# However we want to see correlation in graphical representation so below is function for that
def plot_corr(df, size=11):
    corr = df.corr()
    fig, ax = plt.subplots(figsize=(size, size))
    ax.matshow(corr)
    plt.xticks(range(len(corr.columns)), corr.columns)
    plt.yticks(range(len(corr.columns)), corr.columns)
```

```python
plot_corr(df)
```

```python
df.shape
```

```python
x=df.drop(columns="AHD")
x
```

```python
y=df["AHD"]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
ml=LogisticRegression()
ml.fit(x_train,y_train)
```

```python
y_pred=ml.predict(x_test)
y_pred
```

```python
ml.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
df_cm = pd.DataFrame(cm, index = [i for i in ["1","0"]],
                  columns = [i for i in ["Predict 1","Predict 0"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True)
```

```python
'''cm: This should be a 2D array or matrix (e.g., the output of confusion_matrix() from sklearn).
index: Labels for the rows (1 and 0 for the actual classes).
columns: Labels for the columns (Predict 1 and Predict 0 for predicted classes).
pd.DataFrame: Converts the confusion matrix into a pandas DataFrame, which Seaborn can use for plotting.'''
```

```python
'''Adjusts the figure size to (7, 5) for better visualization.'''
```

```python
'''sns.heatmap: Creates the heatmap.
annot=True: Displays the cell values (confusion matrix counts) on the heatmap.'''
```
