# model Training

```python
x=gold_data_inr[["USD_INR"]]
y=gold_data_inr[["Goldrate"]]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=42)
```

```python
x_train
```

```python
x_train.count()
```

```python
x_test.count()
```

```python
y_train.count()
```

```python
y_test.count()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train_scaled=scaler.fit_transform(x_train)
```

```python
x_test_scaled=scaler.fit_transform(x_test)
```

```python
x_test_scaled
```

```python
x_train_scaled
```

```python
y_train_scaled=scaler.fit_transform(y_train)
```

```python
y_train_scaled
```

```python
y_test_scaled=scaler.fit_transform(y_test)
y_test_scaled
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(x_train_scaled,y_train)
```

```python
regressor.get_params()
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
#y=mx+b m--slope,y--intercept
m=regressor.coef_[0][0]
m
```

```python
b=regressor.intercept_[0]
b
```

```python
import matplotlib as plt
```

```python
import matplotlib.pyplot as plt
# Predictions
x_train_predict = regressor.predict(x_train_scaled)
# Scatter plot of actual data
plt.scatter(x_train, y_train, label="Actual Data")
# Line plot of predictions
plt.plot(x_train, x_train_predict, color="r", label="Predicted Line")
# Labels & legend
plt.xlabel("USD_INR")
plt.ylabel("Goldrate")
plt.legend()
# Show plot
plt.show()
```

```python
x_test_predict=regressor.predict(x_test_scaled)
x_test_predict
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,x_test_predict)
```

# Hyperparameter Optimization

```python
from sklearn.model_selection import RandomizedSearchCV
param_space = {'copy_X': [True,False], 
               'fit_intercept': [True,False], 
               'n_jobs': [1,5,10,15,None], 
               'positive': [True,False]}
```

```python
search=RandomizedSearchCV(regressor,param_space,n_iter=50,cv=5)
```

```python
search.fit(x_train_scaled,y_train)
```

```python
search.best_params_
```

```python
tuned_model=LinearRegression(positive=True,n_jobs= 1,fit_intercept= True ,copy_X= True)
```

```python
tuned_model.fit(x_train_scaled,y_train)
```
