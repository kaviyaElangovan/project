```python

# However we want to see correlation in graphical representation so below is function for that
def plot_corr(df, size=11,label_size=15):
    corr = df.corr()
    fig, ax = plt.subplots(figsize=(size, size))
    ax.matshow(corr)
    plt.xticks(range(len(corr.columns)), corr.columns,rotation=90,fontsize=label_size)
    plt.yticks(range(len(corr.columns)), corr.columns,fontsize=label_size)
```

## Lets defined sigmoid function now and do the math with hand

```python
import math
def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
def predication_function(age):
    z=0.127*age-4.84
    y=sigmoid(z)
    return y
```

```python
age=35
predication_function(age)
```

## 0.402is less than 0.5 which means person with 35 age will not buy insurance

```python
age=47
predication_function(age)
```
